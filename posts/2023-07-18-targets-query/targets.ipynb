```python
def get_json(id):
    headers = {'accept': 'application/json'}
    url = f'https://www.encodeproject.org/files/{id}/?format=json'
    response = requests.get(url, headers=headers)
    encff = response.json()
    return encff
```

```python
def get_slims(id):
    encff = encffs[id]
    system = 'NA'
    cell = 'NA'
    organ = 'NA'
    developmental = 'NA'
    
    try:
        bio_ont = encff['biosample_ontology']
        system_list = bio_ont['system_slims']
        cell_list = bio_ont['cell_slims']
        organ_list = bio_ont['organ_slims']
        developmental_list = bio_ont['developmental_slims']
    
        system = ', '.join(system_list)
        cell = ', '.join(cell_list)
        organ = ', '.join(organ_list)
        developmental = ', '.join(developmental_list)
    except KeyError:
        pass
    
    return [system, cell, organ, developmental]
```

```python
for index, row in df.iterrows():
    system, cell, organ, dev = get_slims(row['identifier'])

    df.at[index, 'system_slims'] = system
    df.at[index, 'cell_slims'] = cell
    df.at[index, 'organ_slims'] = organ
    df.at[index, 'developmental_slims'] = dev
```
